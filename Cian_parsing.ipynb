# Parsing Cian

```python
import os.path

from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://www.cian.ru/cat.php?deal_type=sale&engine_version=2&foot_min=30&metro%5B0%5D=1&offer_type=flat&only_foot=2&p=2&room1=1&room2=1&room3=1&room4=1&room5=1&room6=1&room9=1&sort=creation_date_desc"
```

```python
def get_soup(url):
    driver.get(url)
    return BeautifulSoup(driver.page_source)
```

```python
def get_hrefs(soup):
    divs = soup.find_all("div", class_="_93444fe79c--content--lXy9G")
    ads = set()
    for item in divs:
        href = item.find("a", class_="_93444fe79c--link--eoxce").get("href")
        ads.add(href)
        with open("url_final.txt", "a") as f:
            for i in ads:
                f.write(str(i) + "\n")
```

Так как на циане по фильтру отображается максимум 50 стр, то сделаем перебор ссылок по станция метро, по 5 стр с каждой станции. Записываем в текстовый файл.

```python
for metro_id in range(1, 150):
    for page in range(1, 5):
        soup = get_soup(
            "https://www.cian.ru/cat.php?deal_type=sale&engine_version=2&foot_min=30&metro%5B0%5D="
            + str(metro_id)
            + "&offer_type=flat&only_foot=2&p="
            + str(page)
            + "&room1=1&room2=1&room3=1&room4=1&room5=1&room6=1&room9=1&sort=creation_date_desc"
        )
        get_hrefs(soup)
```

Пройдемся по готовому текстовому файлу и в отдельную папку добавим html страницы всех ссылок.


```python
with open("url_final.txt", "r") as f:
    for line in f:
        print(line)
        filename = line.strip().split("/")[-2]
        if os.path.exists(f"pages/{filename}.html"):
            continue
        else:
            html = get_soup(line.strip())
            with open(f"pages/{filename}.html", "w", encoding="utf-8") as file:
                file.write(str(html))
```
